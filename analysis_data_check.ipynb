```python
import matplotlib.pyplot as plt




# Read the data from old_speed.txt
with open('./analysis_data/speeds.txt', 'r') as file:
    old_speed_data = file.read().strip().split()

# Convert the data to float
old_speed_data = [float(speed) for speed in old_speed_data]

# Read the data from new_speed.txt
with open('speeds.txt', 'r') as file:
    new_speed_data = file.read().strip().split('\n')

# Convert the data to float
new_speed_data = [float(speed) for speed in new_speed_data]

# Create the line plots
plt.figure(figsize=(14, 6))

# Plot for old_speed.txt
plt.subplot(1, 2, 1)
plt.plot(old_speed_data, marker='o', linestyle='-', color='b')
plt.title('Old Speed Data')
plt.xlabel('Index')
plt.ylabel('Speed')
plt.grid(True)

# Plot for new_speed.txt
plt.subplot(1, 2, 2)
plt.plot(new_speed_data, marker='o', linestyle='-', color='g')
plt.title('New Speed Data')
plt.xlabel('Index')
plt.ylabel('Speed')
plt.grid(True)

# Display the plots
plt.tight_layout()
plt.show()
```
